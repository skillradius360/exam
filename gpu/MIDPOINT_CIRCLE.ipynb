```python
import matplotlib.pyplot as plt
r=int(input("Enter the radius"))
x,y=0,r
xcd,ycd=[x],[y]
p=1-r

while x<y:

  if p<0:
    x+=1
    p=p+2*x+1
  else:
    x+=1
    y-=1
    p=p+2*x-2*y+1
  print("x=",x,"y=",y)
  xcd.append(x)
  ycd.append(y)
plt.scatter(xcd,ycd,s=40)
```

```python
import matplotlib.pyplot as plt
r=int(input("Enter the radius"))
x,y=0,r
xcd,ycd=[x],[y]
p=1-r

while x<y:

  if p<0:
    x+=1
    p=p+2*x+1
  else:
    x+=1
    y-=1
    p=p+2*x-2*y+1
  print("x=",x,"y=",y)
  xcd.append(x)
  ycd.append(y)
for i in range(len(ycd)):
  ycd.append(0-i)
for i in range(len(xcd)):
  xcd.append(i)

plt.scatter(xcd,ycd,s=40)
```
