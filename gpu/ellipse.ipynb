```python
import matplotlib.pyplot as plt
import numpy as np

def draw_ellipse(center_x, center_y, a, b, rotation_angle=0):
    theta = np.linspace(0, 2*np.pi, 100)
    cos_theta = np.cos(theta)
    sin_theta = np.sin(theta)

    x = center_x + a * cos_theta * np.cos(rotation_angle) - b * sin_theta * np.sin(rotation_angle)
    y = center_y + a * cos_theta * np.sin(rotation_angle) + b * sin_theta * np.cos(rotation_angle)

    return x, y

def main():
    # Parameters of the ellipse
    center_x = 5
    center_y = 5
    a = 4  # Semi-major axis
    b = 2  # Semi-minor axis
    rotation_angle = np.pi / 4  # Rotation angle (in radians)

    # Generate points of the ellipse
    x, y = draw_ellipse(center_x, center_y, a, b, rotation_angle)

    # Plotting the ellipse
    plt.figure(figsize=(6, 6))
    plt.plot(x, y)
    plt.title('Ellipse Drawing')
    plt.axis('equal')  # Equal scaling of x and y axes
    plt.grid(True)
    plt.show()

if __name__ == '__main__':
    main()
```
