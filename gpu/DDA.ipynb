```python
import matplotlib.pyplot as plt
# m<=1
x1,y1=int(input("Enter the value of x1")),int(input("Enter the value of y1"))
x2,y2=int(input("Enter the value of x2")),int(input("Enter the value of y2"))
dx=abs(x1-x2)
dy=abs(y1-y2)
span = max(dx,dy)
print(dx,dy,span)

x_ord,y_ord=[],[]
x=0
y=0
flag=0
while(flag<span):
    x+=(dx/span)
    y+=(dx/span)
    x_ord.append(x)
    y_ord.append(y)
    flag+=1
    print(x,y)


plt.plot(x_ord,y_ord);
plt.xlabel("x------------>");
plt.ylabel("y----------->");
```
