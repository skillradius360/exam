```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

def original_triangle(ax):
   original_rectangle = Polygon([[0,0],[1,0],[0.5,0.886]],fill=None, edgecolor='r')
   ax.add_patch(original_rectangle)
def final_triangle(ax,dx,dy):
   ver = ([[0,0],[1,0],[0.5,0.886]])
   t_ver=[[v[0]+dx,v[1]+dy] for v in ver]
   f_rectangle = Polygon(t_ver,fill=None, edgecolor='b')
   ax.add_patch(f_rectangle)


def main():
   fig, ax = plt.subplots()
   original_triangle(ax)
   final_triangle(ax,0.5,0.5)

   ax.set_xlim(-0.5, 2)
   ax.set_ylim(-0.5, 2)
   plt.show()
if __name__== "__main__":
    main()
```
